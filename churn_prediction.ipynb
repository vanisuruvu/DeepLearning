```python
#import the lib

import pandas as pd
import numpy as np

# tensorflow lib
import tensorflow
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# to see the tensorflow version

tensorflow.__version__

# load the data

data = pd.read_csv('/content/drive/MyDrive/DeepLearning Afsan/Churn_Modelling (1).csv')
```

```python
data
```

```python
X = data.iloc[: , 3:13]
y = data.iloc[: , 13]

geography = pd.get_dummies(X['Geography'] , drop_first=True)
gender = pd.get_dummies(X['Gender'] , drop_first=True)

X = pd.concat([X , geography, gender], axis = 1)

# drop the duplicate columns

X.drop(columns = ['Geography', 'Gender'] , inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =101 )
```

```python
model = Sequential()
model.add(Dense(180, activation = 'relu'))
model.add(Dense(90, activation = 'sigmoid'))
model.add(Dense(140, activation = 'relu'))
model.add(Dense(20, activation = 'sigmoid'))  # hidden layer 4
model.add(Dense(2))   # output layer
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['Accuracy'])
```

```python
model.fit(X_train, y_train,epochs = 100, validation_data=(X_test, y_test))
```
